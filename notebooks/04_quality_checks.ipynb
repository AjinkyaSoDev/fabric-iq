# 04 — Data Quality Checks

Sanity-checks the Gold layer: row counts, null ratios, value ranges, and FK coverage.

**Prerequisite:** run `03_gold_marts` first.

```python
def read_gold(name):
    return spark.read.format('delta').table(f'gold_{name.lower()}').toPandas()

fv = read_gold('fact_verkeer')
fp = read_gold('fact_waterpeil')
fk = read_gold('fact_waterkwaliteit')
dl = read_gold('dim_locatie')

print('Row counts:')
print(f'  fact_verkeer           {len(fv):>8}')
print(f'  fact_waterpeil         {len(fp):>8}')
print(f'  fact_waterkwaliteit    {len(fk):>8}')
print(f'  dim_locatie            {len(dl):>8}')
```

```python
assert fv['voertuigen_per_uur'].between(0, 6000).all(), 'traffic count out of range'
assert fp['peil_nap_m'].between(-3.5, 1.5).all(), 'water level out of range'
assert fv['locatie_id'].isin(dl['locatie_id']).all(), 'traffic FK violation'
assert fp['locatie_id'].isin(dl['locatie_id']).all(), 'waterpeil FK violation'

print('\n✅ All quality checks passed!')
```
