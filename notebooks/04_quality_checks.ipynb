# 04 — Data quality checks

Sanity-checks the Gold layer: row counts, null ratios, value ranges, and FK coverage.

```python
from pathlib import Path
import pandas as pd
GOLD = Path('..').resolve() / 'lakehouse' / 'gold'

def load(name):
    return pd.read_parquet(GOLD / name / 'part-0.parquet')

fv = load('fact_verkeer'); fp = load('fact_waterpeil'); fk = load('fact_waterkwaliteit')
dl = load('dim_locatie')

print('fact_verkeer        rows:', len(fv))
print('fact_waterpeil      rows:', len(fp))
print('fact_waterkwaliteit rows:', len(fk))
print('dim_locatie         rows:', len(dl))

assert fv['voertuigen_per_uur'].between(0, 6000).all(), 'traffic count out of range'
assert fp['peil_nap_m'].between(-3.5, 1.5).all(), 'water level out of range'
assert fv['locatie_id'].isin(dl['locatie_id']).all(), 'traffic FK violation'
assert fp['locatie_id'].isin(dl['locatie_id']).all(), 'waterpeil FK violation'
print('\nAll quality checks passed ✔')
```
